```python
import os
import torch

from os.path import join as join_path

import numpy as np
from matplotlib import pyplot as plt

from albo.test_functions.synthetic import Gardner1
from albo.test_functions.plotting import get_feasibility_plot_2d
```

```python
def get_traces(dump_dir):
    traces = list()
    for f in os.scandir(dump_dir):
        if f.is_dir() :
            try:
                trace_file = join_path(f.path, 'trace.npy')
                trace = np.load(trace_file, allow_pickle=True)
                traces.append(trace)
            except:
                pass
    return traces
```

```python
test_fcn = Gardner1()
fh, ax = plt.subplots()
contours = get_feasibility_plot_2d(ax, test_fcn)
```

```python
traces = get_traces('/home/stys/work/dumps/albo/gardner1_classic/noiseless_r100_mc500_albest/noiseless_r100_mc500/')
nruns = len(traces)
print('Found %d runs' % nruns)
fh = plt.figure(figsize=(16, 5))
ax1, ax2, ax3 = fh.subplots(1, 3)

true_best_f = - 1.8887

best_points = np.zeros((nruns, test_fcn.dim))    
for i in range(nruns):
    trace = traces[i].tolist()
    idx_best = trace['idx_best']
    idx_best_final = trace['idx_best'][-1]
    x_train = trace['x'].numpy()
    x_best = x_train[idx_best_final]
    best_points[i, :] = x_best
contours = get_feasibility_plot_2d(ax1, test_fcn)
ax1.scatter(best_points[:, 0], best_points[:, 1], c='r', marker='.', zorder=10)
ax1.grid()
ax1.set_xlim([0, 6])
ax1.set_ylim([0, 6])
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Best visited feasible point')

final_points = np.zeros((nruns, test_fcn.dim))
for i in range(nruns):
    trace = traces[i].tolist()
    x_pred = trace['x_al'].numpy()
    final_points[i, :] = x_pred[-1]
contours = get_feasibility_plot_2d(ax2, test_fcn)
ax2.scatter(final_points[:, 0], final_points[:, 1], c='r', marker='.', zorder=50)   
ax2.grid()
ax2.set_xlim([0, 6])
ax2.set_ylim([0, 6])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('ALBO optimum')

n = traces[0].tolist()['y'].shape[0]
y_best = np.zeros((nruns, n))
for i in range(nruns):
    trace = traces[i].tolist()
    idx_best = trace['idx_best']
    y = trace['y'][:, 0]
    valid = (idx_best >= 0)
    y_best[i, :] = np.where(valid, y[idx_best], np.inf)
y_best_median = np.median(y_best, 0) - true_best_f
every_3 = np.array(list(x % 3 == 0 for x in range(len(valid))))
y_best_q75 = np.where(every_3, np.quantile(y_best - true_best_f, 0.75, 0), 0)
y_best_q25 = np.where(every_3, np.quantile(y_best - true_best_f, 0.25, 0), 0)
ax3.errorbar(
    range(n),
    y_best_median,
    np.array([y_best_median - y_best_q25, y_best_q75 - y_best_median]),
    linewidth=3,
    elinewidth=1,
    capsize=3,
    label='best visited feasible value'
)
ax3.grid()

L = np.zeros((nruns, n))
for i in range(nruns):
    for j, t in enumerate(traces[i].tolist()['traces_inner']):
        if j < n:
            if t is not None:
                L[i, j] = - t[-1]['L']
            else:
                L[i, j] = np.nan
L_median = np.median(L, axis=0) - true_best_f
every_3_1 = np.array(list((x+1) % 2 == 0 for x in range(len(valid))))
L_q75 = np.where(every_3_1, np.quantile(L - true_best_f, 0.75, 0), 0)
L_q25 = np.where(every_3_1, np.quantile(L - true_best_f, 0.25, 0), 0)
ax3.errorbar(
    range(n),
    L_median,
    np.array([L_median - L_q25, L_q75 - L_median]),
    linewidth=3,
    elinewidth=1,
    capsize=3,
    label='AL prediction'
)

pesc_results = np.load('./pesc_task/best_f_visited.npy')
pesc_median = np.nanmedian(pesc_results - true_best_f, axis=0)[:n]
every_3_2 = np.array(list((x+2) % 2 == 0 for x in range(len(valid))))
pesc_q75 = np.where(every_3_2, np.nanquantile(pesc_results - true_best_f, q=0.75, axis=0)[:n], 0)
pesc_q25 = np.where(every_3_2, np.nanquantile(pesc_results - true_best_f, q=0.25, axis=0)[:n], 0)

ax3.errorbar(
    range(n),
    pesc_median,
    np.array([pesc_median - pesc_q25, pesc_q75 - pesc_median]),
    linewidth=3,
    elinewidth=1,
    capsize=3,
    linestyle='--',
    label='PESc'
)

# last estimated point in PESc is broken
pesc_est = np.load('./pesc_task/best_f_estimated.npy')
pesc_est_median = np.nanmedian(pesc_est - true_best_f, axis=0)[:n-1]
every_3_2 = np.array(list((x+2) % 2 == 0 for x in range(n-1)))
pesc_est_q75 = np.where(every_3_2, np.nanquantile(pesc_est - true_best_f, q=0.75, axis=0)[:n-1], 0)
pesc_est_q25 = np.where(every_3_2, np.nanquantile(pesc_est - true_best_f, q=0.25, axis=0)[:n-1], 0)

ax3.errorbar(
    range(n-1),
    pesc_est_median,
    np.array([pesc_est_median - pesc_est_q25, pesc_est_q75 - pesc_est_median]),
    linewidth=3,
    elinewidth=1,
    capsize=3,
    linestyle='--',
    label='PESc prediction'
)


ax3.legend()
ax3.set_xlabel('blackbox evaluations')
ax3.set_title('Log utility gap')

ax3.set_xlim([0, 50])
ax3.set_ylim([0, 1])

# check log scale
ax3.set_yscale('log')
ax3.set_ylim([0.00001, 2])
```
